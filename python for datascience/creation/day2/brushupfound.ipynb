```python
class stack:
    def __init__(self):
        self.data=[]
    def push(self,x):
        self.data.append(x)
        print(f"{self.data} is pushed")
    def pop(self):
        if self.is_empty():
            raise IndexError("Pop from empty stack")
        return self.data.pop()
    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.data[-1]
    def is_empty(self):
        return len(self.data)==0
    def size(self):
        return len(self.data)
```

```python
s=stack()
s.push(1)
s.push(4)
assert s.pop()==4
assert s.pop()==1
```
