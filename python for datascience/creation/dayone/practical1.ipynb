```python
#Write a Python program to check if a given number is prime.
no=int(input("enter a no:"))
def prime(no):
    a=2
    if no<2:
        return False
    while(a<no):
        if no%a==0:
            return False
        a+=1
    return True
if prime(no):
    print(f"{no} is the prime no.")
else:
    print(f"{no} is not the prime no.")
```

```python
# Create a function that converts a list of temperatures from Celsius to Fahrenheit.
def conversion_fahren(temp=[]):
    i=0
    for fahren in temp:
        temp[i]=(fahren*(9/5))+32
        i+=1
    return temp

temp_celc=[13,37,23,79]
temperature=conversion_fahren(temp_celc)
for degree in temperature:
    print(f"{degree:.2f} fahrenheit")
print(temperature)

```

```python
# Implement a function to find the factorial of a number using recursion.
def fact(no,a=1):
    if no==0 or no==1:
        return a
    else:
        return fact(no-1,no*a)
factorial=fact(15)
print(factorial)
```

```python
# Create a function that takes a list of numbers and returns only the even numbers.
def even_only(no=[]):
    temp=[]
    for i in no:
        if i%2==0:
            temp.append(i)
    return temp
numbers=[]
n=int(input("enter the total no of list"))
for i in range(n):
    numbers.append(int(input("enter the no:")))
print(even_only(numbers))
```

```python
# Implement a class Circle with a method to calculate the area and circumference.
class Circle:
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.143*self.r**2
    def circumference(self):
        return 2*3.143*self.r
c1=Circle(int(input("enter hte radius")))
print(f"the radius is {c1.r}\n the area is {c1.area()} \n the circumference is {c1.circumference()}")
```

```python
#Implement a function to remove duplicates from a list without using set().
def duplicate_remove(aList):
    temp=[]
    for element in aList:
        if element not in temp:
            temp.append(element)
    return temp
xList=[1,2,3,41,3,2]
result=duplicate_remove(xList)
print(result)
```

```python
# Write a Python program to find the largest and smallest numbers in a list.
def separator(aList):
    smallest=aList[0]
    largest=aList[0]
    for element in aList:
        if smallest> element:
            smallest=element
        if largest<element:
            largest=element
    return smallest,largest
no=[1,2,4,-343,-342,31,24,5465,6,4567,574,2,10,4]
smallest,largest=separator(no)
print(smallest,largest)
```

```python
# Create a class Rectangle with methods to calculate area and perimeter.
class Rectangle:
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l*self.b
    def perimeter(self):
        return 2*(self.l+self.b)
print("enter the length and breadth:")
a=int(input())
b=int(input())
c1=Rectangle(a,b)
print(f"the length and breadth are ({a} {b})\n the area is {c1.area()} \n the circumference is {c1.perimeter()}")
```

```python
# Write a function to reverse a list without using the built-in reverse() method.
def list_reverse(Alist):
    temp=[]
    no=len(Alist)
    for i in range(1,no+1):
        temp.append(Alist[-i])
    return temp
xList=[1,2,5,8,"aakash"]
print(list_reverse(xList))
```

```python
# Implement a Python function that generates the Fibonacci sequence up to n terms
def fibonacci_seq(n):
    if n==0:
        return []
    elif n==1:
        return [0]
    elif n==2:
        return [0,1]
    else:
        seq=fibonacci_seq(n-1)
        seq.append(seq[-1]+seq[-2])
        return seq
print(fibonacci_seq(10))
```

```python
# 17.Create a function that accepts a variable number of arguments and returns their sum.
def sum(*args):
    total=0
    for no in args:
        total+=no
    return total
print(sum(2,3,4))
```

```python
# 18.Implement a class Employee with attributes name and salary and a method to apply a 10% raise
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def apply_raise(self):
        self.salary = self.salary * 1.1  # 10% raise
        return self.salary

e1 = Employee("Alice", 5000)
print(f"Before raise: {e1.salary}")
print(f"After 10% raise: {e1.apply_raise()}")
```

```python
# Write a Python program to check if two strings are anagrams of each other.
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

s1 = "listen"
s2 = "silent"
print(f"'{s1}' and '{s2}' are anagrams? {are_anagrams(s1, s2)}")
```

```python
# 20.Create a function that calculates the sum of digits of a given number.
def sum_of_digit(numbers):
    no=str(numbers)
    a=0
    for element in no:
        a+=int(element)
    return a
a=sum_of_digit(1233)
print(a)
```
