```python
# create a table for books with fields:title,author,year
# insert 3 books and query by author.
# update a book's year
# delete a book
```

```python
from sqlalchemy import create_engine
engine=create_engine('sqlite:///bookdb.db',echo=True)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer,String

Base=declarative_base()
class Book(Base):
    __tablename__='Books'
    title=Column(String,primary_key=True)
    author=Column(String)
    year=Column(Integer)
#create the table
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session=sessionmaker(bind=engine)
Session=Session()
```

```python
book1=Book(title='reverend insanity',author='gu zen re',year=2000)
book2=Book(title='swallowed star',author='ieattomato',year=2010)
book3=Book(title='lotm',author='ieattomato',year=2001)
Session.add(book1)
Session.add(book2)
Session.add(book3)
```

```python
# books=Session.query(Book).all()
books = Session.query(Book).filter_by(author='ieattomato').all()
for b in books:
    print(b.title,b.author,b.year)

```

```python
books=Session.query(Book).filter_by(title='reverend insanity').first()
books.year=2065
# Session.commit()
```
