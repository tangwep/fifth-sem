```python
import numpy as np 
import matplotlib.pyplot as plt 

X = np.arange(1,10,2)
y = X**2
z = X**3


plt.plot(X, y, label= "X Squared")
plt.plot(X, z, label="X cubed")
plt.xlabel("X")
plt.ylabel("y")
plt.title("X vs y")
plt.legend()
plt.savefig("plot.png")
```

```python
# WAP to count the frequency of characters in a sentence
# sentence = "ABRACADABRA"
sentence = input("Enter a sentence")
print("The entered sentence is ",sentence)
aDict = {}
for letter in sentence:
    if letter not in aDict:
        aDict[letter] = 1
    else:
        aDict[letter] += 1
print(aDict)
```
