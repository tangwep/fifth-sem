```python
#*args it takes input as tuple, whien we dont know the num of arguments
def average(*args):
    s=0
    for arg in args:
        s=arg+s
    return s/len(args)
average(1,2,3,4,5)
```

```python
# **args 9t is used for dictionarys
def func(**kwargs):
    for item in kwargs.items():
        print(item)
func(name="ram",address="dharan")
```

```python
#merge of args and kwargs
def myfunc(*args,**kargs):
    print(args)
    print(type(kargs))
    print(kargs)
myfunc(2,3,a=10,b=20)#it is position fixed from he functions itself
```

```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age 
    def __str__(self):
        return f'Animal{self.name} is {self.age} years old'
    def speak(self):
        print(f"{self.name} is speaking!")
    def __repr__(self):
        return f'Animal:{self.name}  |Age: {self.age}'

a1=Animal("Lucky",12)
a1
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f'({self.x},{self.y})'
    def __add__(self,other):
        new_x=self.x+other.x
        new_y=self.y+other.y
        return Point(new_x,new_y)
p1=Point(2,3)
p2=Point(3,2)
p3=p1+p2
print(p3)
```

```python
#wap to creat a class named distance that has feet and inches as data members. overload any 2 operators
class Distance:
    def __init__(self,feet,inch):
        self.feet=feet
        self.inch=inch
    def __str__(self):
        return f'The distance is {self.feet} feet and {self.inch} inch'
    def __add__(self,other):
        new_feet=self.feet+other.feet
        new_inch=self.inch+other.inch
        if new_inch>12:
            new_feet+=new_inch//12
            new_inch=new_inch%12
        return Distance(new_feet,new_inch)
    def __sub__(self,other):
        if self.feet<other.feet:
            print("the distance is negative.")
        else:
            new_feet=self.feet-other.feet
            new_inch=self.inch-other.inch
            if new_inch<0:
                new_feet-=1
                new_inch=new_inch+12
            return Distance(new_feet,new_inch)
d1=Distance(2,2)
d2=Distance(4,11)
d3=d1+d2
d4=d1-d2
print(d1)
print(d2)   
print("the new distances are:")
print(d3)
print(d4)
```
