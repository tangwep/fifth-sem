```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
```

```python
db=client["school"]
collection=db["students"]
#insert a single document
student={"name":"Alice","age":25,"subject":["maths","science"]}
collection.insert_one(student)
print(client.list_database_names())
```

```python
#insert a singel documents
student={"name":"Alice","age":25,"subjects":["Math","science"]}
collection.insert_one(student)
```

```python
many_students=[
    {"name": "Ram", "age":18, "subjects":["Math", "Science"]},
    {"name": "Shyam", "age":27, "subjects":["Math", "Science"]}
]
collection.insert_many(many_students)
```

```python
# Query to display all students
for data in collection.find():
    print(data)
```

```python
#query to display conditional results
for data in collection.find({"age":{"$gt":20}}):
    print(data)
```

```python
print(db.list_collection_names())

collection.drop()

print(db.list_collection_names())
```

```python
print(client.list_database_names())
client.drop_database("school")
print(client.list_database_names())
```
