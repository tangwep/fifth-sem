```python
import requests
from pprint import pprint
datas=[]
api_key = "1e679f6aa64bfb3ba0eabd0dff937661"
city=["biratnagar","dharan","pokhara","kathmandu","lumbini"]
for city_name in city:
    url=f"http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}&units=metric"
    response=requests.get(url)
    if response.status_code==200:
        temp_data= response.json()
        datas.append(temp_data)
        pprint(datas)
    else:
        print(response)
```

```python
import pandas as pd
from pandas import json_normalize
for data in datas:    
    df=json_normalize(data)
    df_weather=df[[
        'name',
        'main.temp',
        'main.humidity',
        'wind.speed',
        'weather',
        'main.feels_like'
    ]].copy()  #copy is done so to creat a new dataframe
    # print(df_weather.to_string())
    df_weather['Description']=df['weather'].apply(lambda x:x[0]['description'])
    df_weather=df_weather.drop(columns=['weather'])
    pprint(df_weather)
    
```
