```python
import requests
import pandas as pd 
from pandas import json_normalize

api_key = "1e679f6aa64bfb3ba0eabd0dff937661"
cities=["Dharan", "Biratnagar", "Kathmandu", "Itahari", "Pokhara"]
df_list = []
for city in cities:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        df = json_normalize(data)
        df_weather = df[[
            'name',
            'main.temp',
            'main.humidity',
            'main.feels_like',
            'wind.speed',
            'weather'
        ]].copy()

        df_weather['Description'] = df['weather'].apply(lambda x:x[0]['description'])
        df_weather = df_weather.drop(columns=['weather'])
        df_list.append(df_weather)

final_df = pd.concat(df_list, ignore_index=True)
final_df.to_csv("weather_data.csv")
```
