```python
import pandas as pd
df=pd.DataFrame({
    "EID":[101,102,103,104],
    "Name":["Alice","Bob","Charlie","David"],
    "Level":["Junior","senior","Junior","Mid"]
})
print(df.info())
```

```python
#convert level column to categorial data type
df['Level']=df['Level'].astype('category')
print(df.info())
```

```python
df["Level"]=df['Level'].cat.rename_categories({
    "Junior":"Jr",
    "senior":"Sr"
})
print(df.head())
```

```python
df["Level_code"]=df['Level'].cat.codes
print(df.head())
```

```python
encoded=pd.get_dummies(df['Level'],prefix='Lev',dtype=int)
print(encoded)
```

```python
df_final=pd.concat([df,encoded],axis=1)
print(df_final.head())
```
