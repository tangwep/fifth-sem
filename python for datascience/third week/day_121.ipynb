```python
import pandas as pd
df=pd.DataFrame({
    'Date':['2023-05-01','2023-09-02','2023-08-03','2023-03-04'],
    'Sales':[250,300,400,500]
})
print(df)
```

```python
print(df.info())
```

```python
df['Date']=pd.to_datetime(df['Date'])
print(df.info())
```

```python
df=df.set_index('Date')
df.head()
```

```python
df['Year']=df.index.year
df['Month']=df.index.month
df['Day']=df.index.day
df['Weekday']=df.index.day_name()
df['Month_Name']=df.index.month_name()
print(df.head())
```

```python
df.loc["2023-05"]
```

```python
df.plot.bar(x='Weekday',y='Sales',title='sales by day name')
```

```python
df.to_csv("file.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=pd.Series([0,1,10,100,1000])
x_log=np.log10(x)
print(x_log)
plt.plot(x,label='x')
plt.plot(x_log,label='x_log')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
X=np.array([[10],[200],[800],[900],[1000]])
scaler=MinMaxScaler(feature_range=(-1,1))
X_scaled=scaler.fit_transform(np.array(X).reshape(-1,1))    
print(X_scaled)
print(X)
#The values are scaled between 0 and 1
```

```python
X=[10,200,800,900,1000]
min=min(X)
max=max(X)
for i in range(len(X)):
    X[i]=(X[i]-min)/(max-min)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
X=np.array([[10],[200],[800],[900],[1000]])
scaler=StandardScaler()
X_scaled=scaler.fit_transform(np.array(X).reshape(-1,1))    
print(X_scaled)
print(X)
#The values are scaled between 0 and 1
```
