```python
import pandas as pd
```

```python
import random
import tqdm
names=['ram','shyam','hari','krishna','sita']
final_list=[]
for i in tqdm.tqdm(range(1,10000)):
    random_name=names[random.randint(0,4)]
    final_list.append({
        "EID":i,
        "Name":random_name,
        "Salary":100000*random.random()
    })
    df=pd.DataFrame(final_list)
    df.to_csv("big_data.csv")
```

```python
first_chunk=True
output_file='processed_data.csv'
for chunk in pd.read_csv('big_data.csv',chunksize=100):
    chunk['Gender']=chunk['Name'].apply(lambda x:'Female' if x in "sita" else 'M')
    chunk.to_csv(output_file,
                 index=False,
                 mode='w' if first_chunk else 'a',
                 header=first_chunk)
    first_chunk=False
```
